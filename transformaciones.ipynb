```python
import pandas as pd
import json
import ast 
#importar dataset
movies=pd.read_csv('movies_dataset.csv',sep=',')
movies.head()
```

```python
def convertir1(object):  ## funciona con genres, 
    Lista = []
    for i in ast.literal_eval(object):
        Lista.append(i['name'])
    return Lista

movies["genres"]=movies["genres"].apply(convertir1) #desanidamos genres y extraemos names
```

```python
#esta funcion sirve para lista de diccionarios con string
def convertir2(obj): 
    
    if isinstance(obj, str) and '{' in obj:
        dic = ast.literal_eval(obj)

        return dic['name']
## desanido bellongs
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(convertir2)
```

```python
#desanidar production_companies
def convertir3(obj): 
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
movies['production_companies'] = movies['production_companies'].apply(convertir3)
```
